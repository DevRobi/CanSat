```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import socket
from time import sleep


# Define the number of data points to keep track of
NUM_POINTS = 200
data_points = np.zeros((NUM_POINTS, 3))
# Initialize the plot
fig = plt.figure()
ax1 = plt.subplot2grid((2,1),(0,0))
ax1.set_aspect('equal')
ax2 = plt.subplot2grid((2,1),(1,0))

def data_gen():
	global data_points
	data = [1,2,3]


	data_points = np.roll(data_points, -1, axis=0)
	data_points[-1, :] = data
	yield data_points


def init():
	ax1.cla()
	limits=200
	ax1.set_xlim(-5,5)
	ax1.set_ylim(-5,5)

def run(data):
	print(data)
	# update the data
	# Draw the line connecting the data points
	ax1.plot(data_points[:, 0], data_points[:, 1])
   
	ax2.scatter(data_points[:, 0], data_points[:, 1])
	
	
	ax1.set_title("példa ábra")
	ax1.grid()
	# ax.figure.canvas.draw()


# Only save last 100 frames, but run forever
ani = animation.FuncAnimation(fig, run, data_gen, interval=500, init_func=init,cache_frame_data=False)						   
plt.show()
```

```python
time = []
alt = []
pressure = []
temp = []
file = open('log.txt', 'r')
for line in file.readlines():
    if "bytearray" not in line:
        # Parse the string as a tuple
        tuple_object = ast.literal_eval(line)

# Convert the tuple to a list
        list_object = list(tuple_object)
        
        if list_object[6] >= 112750 and list_object[6] <= 112922:
        
            #convert time value to seconds only
            t = list_object[6]
            h = int(str(t)[:2]) * 3600
            m = int(str(t)[2:4]) * 60
            s = int(str(t)[4:])
            time.append(h+m+s - (11*3600 + 28 * 60 + 50))
            alt.append((1 - math.pow((list_object[2] / standardPressure), 0.190284)) * 145366.45 * 0.3048 - 140)
            pressure.append(list_object[2])
            temp.append(list_object[0])

plt.scatter(time, alt)
# add labels and title
plt.xlabel('Time / s')
plt.ylabel('Altitude / m')
plt.title('Time vs Altitude (from ground level) Graph Test April 2')

# display the plot
plt.show()
```
